```python
import gensim
from gensim.models.doc2vec import TaggedDocument, Doc2Vec
import pickle
import csv 
import smart_open
```

```python
def read_corpus(fname, tokens_only=False):
    with smart_open.open(fname, encoding="ASCII") as f:
        for i, line in enumerate(f):
            words = line.split()

            if tokens_only:
                yield words
            else:
                yield TaggedDocument(words, [i])
```

```python
malware_train = "/home/haohao/Downloads/Malware Modus Operandi/Malware_train"
```

```python
benign_train = "/home/haohao/Downloads/Malware Modus Operandi/Benign_train"
```

```python
save_path = "/home/haohao/Downloads/Malware Modus Operandi/Models/"
```

# Create benign & malware corpus

```python
benign_corpus = list(read_corpus(benign_train))
```

```python
print(benign_corpus[:2])
```

```python
malware_corpus = list(read_corpus(malware_train))
```

```python
print(malware_corpus[:2])
```

# Save tagged doc

```python
with open(save_path + "benign_taggedDoc.pkl", 'wb') as f:
    pickle.dump(benign_corpus, f)
```

```python
with open(save_path + "malware_taggedDoc.pkl", 'wb') as f:
    pickle.dump(malware_corpus, f)
```

# Build Doc2Vec model

```python
model = Doc2Vec(vector_size=400, alpha=0.075, window=1, min_count=2, epochs=20)
```

```python
model.build_vocab(benign_corpus)
```

```python
model.build_vocab(malware_corpus)
```

```python
model.train(benign_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.train(malware_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

# Save models

```python
model.save(save_path + "Doc2Vec_Benign")
```

```python
model.save(save_path + "Doc2Vec_Malware")
```

# Get feature vectors

```python
def get_feature_vectors(model, taggeddocs, fname):
    with open(fname, 'w', newline='') as f:
        writer = csv.writer(f)
        for doc_id in range(len(taggeddocs)):
            # infer the vector of a doc, taggeddocs[doc_id].words is the list of words in the doc
            vector = model.infer_vector(taggeddocs[doc_id].words)
            writer.writerow(vector)
```

```python
get_feature_vectors(model, benign_corpus, save_path + "benign_train_feature_vectors.csv")
```

```python
get_feature_vectors(model, malware_corpus, save_path + "malware_train_feature_vectors.csv")
```
